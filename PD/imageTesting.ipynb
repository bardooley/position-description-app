```python
school_name = "Duxbury High School"
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# Search for the school image using Google Custom Search API
# Note: You'll need to set up a Google Custom Search API key and Search Engine ID


search_url = f"https://www.googleapis.com/customsearch/v1"
params = {
    'key': 'AIzaSyCcSaOIp7HM2LGUMnNwaDvisF2BR0D1HN4',  # Replace with your API key
    'cx': '752267de5b1734470',  # Replace with your Search Engine ID
    'q': f"{school_name} + logo",
    'searchType': 'image',
    'num': 1
}

response = requests.get(search_url, params=params)
if response.status_code == 200:
    data = response.json()
    if 'items' in data:
        image_url = data['items'][0]['link']
        
        # Download and display the image
        img_response = requests.get(image_url)
        img = Image.open(BytesIO(img_response.content))
        
        plt.figure(figsize=(10, 8))
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Image of {school_name}")
        plt.show()
    else:
        print("No images found")
else:
    print("Error searching for images")


```
